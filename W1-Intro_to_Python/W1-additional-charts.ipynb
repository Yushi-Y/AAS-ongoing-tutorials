### Additional Charts


This notebook provides a few more examples of common probability distributions with charts.

#### Exponential

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import math
```

```python
exp_rvs = stats.expon.rvs(scale = 1, size = 1000)
exp_x = np.linspace(exp_rvs.min(), exp_rvs.max(), 100)
exp_y = stats.expon.pdf(x = exp_x, scale = 1)

plt.figure()
plt.hist(exp_rvs, density=True, color = "#67a9cf", label = "Empirical")
plt.plot(exp_x, exp_y, color = "#ef8a62", label = "Theoretical")
plt.legend(loc = "upper right")
plt.xlabel("Value")
plt.ylabel("Distribution")
plt.title("Exponential ")
plt.show()
```

#### Binomial

```python
n1 = 100
n2 = 1000
p1 = .3
p2 = .5 

binom2_rvs = stats.binom.rvs(p = p2, n = n2, size = 1000)
binom2_x = np.unique(binom2_rvs)
binom2_y = np.array([stats.binom.pmf(p = p2, n = n2, k = x) for x in binom2_x])

binom1_rvs = stats.binom.rvs(p = p1, n = n1, size = 1000)
binom1_x = np.unique(binom1_rvs)
binom1_y = np.array([stats.binom.pmf(p = p1, n = n1, k = x) for x in binom1_x])

fig, ax = plt.subplots(1,2, figsize = (10,5))
ax[0].hist(binom1_rvs, bins = binom1_rvs.max() - binom1_rvs.min(), density=True, color = "#67a9cf", label = "Empirical")
ax[0].plot(binom1_x, binom1_y, color = "#ef8a62", label = "Theoretical")
ax[0].legend(loc = "upper right")
ax[0].set_xlabel("Value")
ax[0].set_ylabel("Distribution")
ax[0].set_title(f"Binomial({n1}, {p1})")

ax[1].hist(binom2_rvs, bins = binom2_rvs.max() - binom2_rvs.min(), density=True, color = "#67a9cf", label = "Empirical")
ax[1].plot(binom2_x, binom2_y, color = "#ef8a62", label = "Theoretical")
ax[1].legend(loc = "upper right")
ax[1].set_xlabel("Value")
ax[1].set_ylabel("Distribution")
ax[1].set_title(f"Binomial({n2}, {p2})")

plt.tight_layout()

plt.savefig('binomial_pmf')
plt.show()
```

#### Normal

```python
m1 = 10
m2 = 0
m3 = 0
s1 = 5
s2 = 2
s3 = 1

norm3_rvs = stats.norm.rvs(loc = m3, scale = s3, size = 1000)
norm3_x = np.linspace(norm3_rvs.min(), norm3_rvs.max(), 100)
norm3_y = stats.norm.pdf(x = norm3_x, loc = m3, scale = s3)

norm2_rvs = stats.norm.rvs(loc = m2, scale = s2, size = 1000)
norm2_x = np.linspace(norm2_rvs.min(), norm2_rvs.max(), 100)
norm2_y = stats.norm.pdf(x = norm2_x, loc = m2, scale = s2)

norm1_rvs = stats.norm.rvs(loc = m1, scale = s1, size = 1000)
norm1_x = np.linspace(norm1_rvs.min(), norm1_rvs.max(), 100)
norm1_y = stats.norm.pdf(x = norm1_x, loc = m1, scale = s1)

fig, ax = plt.subplots(1,3, figsize = (10,5))
ax[0].hist(norm1_rvs, density=True, color = "#67a9cf", label = "Empirical")
ax[0].plot(norm1_x, norm1_y, color = "#ef8a62", label = "Theoretical")
ax[0].legend(loc = "upper right")
ax[0].set_xlabel("Value")
ax[0].set_ylabel("Distribution")
ax[0].set_title(f"Normal({m1}, {s1})")

ax[1].hist(norm2_rvs, density=True, color = "#67a9cf", label = "Empirical")
ax[1].plot(norm2_x, norm2_y, color = "#ef8a62", label = "Theoretical")
ax[1].legend(loc = "upper right")
ax[1].set_xlabel("Value")
ax[1].set_ylabel("Distribution")
ax[1].set_title(f"Normal({m2}, {s2})")

ax[2].hist(norm3_rvs, density=True, color = "#67a9cf", label = "Empirical")
ax[2].plot(norm3_x, norm3_y, color = "#ef8a62", label = "Theoretical")
ax[2].legend(loc = "upper right")
ax[2].set_xlabel("Value")
ax[2].set_ylabel("Distribution")
ax[2].set_title(f"Normal({m3}, {s3})")

plt.tight_layout()

plt.savefig('normal_pdf')
plt.show()
```

#### Repeated Coinflips

```python
binom_rvs = stats.binom.rvs(p = .5, n = 10, size = 10)

plt.figure()
plt.hist(binom_rvs, bins = 10, range=(0,10), color = "#67a9cf")
plt.yticks(np.array(range(5)))
plt.title('Number of Heads in Ten Flips')
plt.xlabel('Count')
plt.ylabel('Frequency')
plt.savefig('coin_flip_experiment')
plt.show()
```

```python
n = 1000
p = .5

bern_rvs = stats.bernoulli.rvs(p = p, size = n)
bern_rvs2 = stats.bernoulli.rvs(p = p, size = (n,n))
bern_rvs2 = np.mean(bern_rvs2,axis=1)
bern_x = np.linspace(bern_rvs2.min(), bern_rvs2.max(), 100)
bern_y = stats.norm.pdf(x = bern_x, loc = p, scale = np.sqrt(p*(1-p))/np.sqrt(n))

uni_rvs = stats.uniform.rvs(size = n)
uni_rvs2 = stats.uniform.rvs(size = (n,n))
uni_rvs2 = np.mean(uni_rvs2 ,axis=1)
uni_x = np.linspace(uni_rvs2.min(), uni_rvs2.max(), 100)
uni_y = stats.norm.pdf(x = uni_x, loc = p, scale = np.sqrt(1/12)/np.sqrt(n))


geo_rvs = stats.geom.rvs(p=.5, size = n)
geo_rvs2 = stats.geom.rvs(p=.5, size = (n,n))
geo_rvs2 = np.mean(geo_rvs2 ,axis=1)
geo_x = np.linspace(geo_rvs2.min(), geo_rvs2.max(), 100)
geo_y = stats.norm.pdf(x = geo_x, loc = 1/p, scale = np.sqrt((1-p)/(p**2))/np.sqrt(n))


fig, ax = plt.subplots(3,2,figsize=(15,10))

ax[0][0].hist(bern_rvs, color = "#67a9cf")
ax[0][0].set_title(f'Sampling of {n} Variables')
ax[0][0].set_xlabel('Value')
ax[0][0].set_ylabel('Frequency')

ax[0][1].hist(bern_rvs2, density=True, color = "#67a9cf", label = "Empirical")
ax[0][1].plot(bern_x, bern_y, color = "#ef8a62", label = "Theoretical")
ax[0][1].set_title(f'Means of {n} Samples of {n} Variables')
ax[0][1].legend(loc='upper right')
ax[0][1].set_xlabel('Mean')
ax[0][1].set_ylabel('Frequency')

ax[1][0].hist(uni_rvs, color = "#67a9cf")
ax[1][0].set_xlabel('Value')
ax[1][0].set_ylabel('Frequency')

ax[1][1].hist(uni_rvs2, density=True, color = "#67a9cf", label = "Empirical")
ax[1][1].plot(uni_x, uni_y, color = "#ef8a62", label = "Theoretical")
ax[1][1].legend(loc='upper right')
ax[1][1].set_xlabel('Mean')
ax[1][1].set_ylabel('Frequency')


ax[2][0].hist(geo_rvs, color = "#67a9cf")
ax[2][0].set_xlabel('Value')
ax[2][0].set_ylabel('Frequency')

ax[2][1].hist(geo_rvs2, density=True, color = "#67a9cf", label = "Empirical")
ax[2][1].plot(geo_x, geo_y, color = "#ef8a62", label = "Theoretical")
ax[2][1].legend(loc='upper right')
ax[2][1].set_xlabel('Mean')
ax[2][1].set_ylabel('Frequency')

ax[0][0].annotate('Bernoulli', xy=(0, 0.5), xytext=(-ax[0][0].yaxis.labelpad - 5, 0),
                xycoords=ax[0][0].yaxis.label, textcoords='offset points',
                size='large', ha='right', va='center')

ax[1][0].annotate('Uniform', xy=(0, 0.5), xytext=(-ax[1][0].yaxis.labelpad - 5, 0),
                xycoords=ax[1][0].yaxis.label, textcoords='offset points',
                size='large', ha='right', va='center')

ax[2][0].annotate('Geometric', xy=(0, 0.5), xytext=(-ax[2][0].yaxis.labelpad - 5, 0),
                xycoords=ax[2][0].yaxis.label, textcoords='offset points',
                size='large', ha='right', va='center')

fig.tight_layout()
plt.savefig('clt_demos')
plt.show()
```

```python
n = 100
count = 15

die_rolls = stats.randint(1,7).rvs((n,count))
index = np.ones((n,count))
cum_rolls = die_rolls.cumsum(axis=0)
index = index.cumsum(axis=0)
cum_rolls = cum_rolls/index

plt.figure()
plt.plot(cum_rolls, color = "#67a9cf")
plt.title(f'Mean of {n} Die Rolls')
plt.xlabel('Cumulative Rolls')
plt.ylabel('Sample Average')
fig.tight_layout()
plt.savefig('lln_demos')
plt.show()
```
